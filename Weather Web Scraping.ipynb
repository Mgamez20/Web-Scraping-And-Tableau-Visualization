```python
import requests
```

```python
#URL página del clima
url  = "https://world-weather.info/forecast/usa/chicago/march-2021/"
```

```python
#Obtención del contenido de la página
content = requests.get(url).content
```

```python
content
```

```python
#Estructura de la información
from bs4 import BeautifulSoup
soup = BeautifulSoup(content, 'html.parser')
soup
```

```python
#Obtención del clima y el día 
weather = []
day = []
for info in BeautifulSoup(content).find_all('ul', {'class':'ww-month'}):
    for columns in info.find_all('span'):
        weather.append(columns.text.strip())
        
    for columns in info.find_all('div'):
        day.append(columns.text.strip())
        
    
```

```python
weather
```

```python
day
```

```python
import pandas as pd
```

```python
#Creación de dataframe para contener la Fecha y el Clima
df_weather = pd.DataFrame(columns = ['Date', 'Weather'])
```

```python
#Formato de fecha 
weather_dates = []
for number in day:
    if int(number)<10:
        weather_dates.append('0'+str(number)+'/03/2021')
    else:
        weather_dates.append(str(number)+'/03/2021')
```

```python
weather_dates
```

```python
#Conversión de datos a Series
pd.Series(weather)
pd.Series(weather_dates)
```

```python
#Agregación de datos al DataFrame
df_weather['Date'] = weather_dates
df_weather['Weather'] = weather
```

```python
df_weather
```

```python
#Creación de archivo CSV para el DataFrame
df_weather.to_csv("weather_chicago_march2021.csv")
```
